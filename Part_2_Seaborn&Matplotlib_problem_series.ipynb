```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

1. Load dataset Iris

```python
# load the iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

2. Plot a histogram of the sepal length of the iris flowers.

```python
plt.figure(figsize=(6,4))
sns.histplot(iris['sepal_length'], bins=20, kde=True)
plt.title("Histogram of Sepal Length")
plt.show()
```

3. Plot a bar chart of the mean values of each of the four features for each species.

```python
plt.figure(figsize=(8,6))
sns.barplot(data=iris.melt(id_vars="species"), x="variable", y="value", hue="species", ci=None)
plt.title("Mean of Features by Species")
plt.show()
```

4. Create a scatter plot of the sepal length and width, with different colors for different species.

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
plt.title("Sepal Length vs Width by Species")
plt.show()
```

5. Create a boxplot of the petal width for each species.

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=iris, x="species", y="petal_width")
plt.title("Boxplot of Petal Width by Species")
plt.show()
```

6. Create a violin plot of the sepal length, with different colors for different species.

```python
plt.figure(figsize=(6,4))
sns.violinplot(data=iris, x="species", y="sepal_length")
plt.title("Violin Plot of Sepal Length by Species")
plt.show()
```

7. Create a violin plot of the sepal  width, with different colors for different species.

```python
plt.figure(figsize=(6,4))
sns.violinplot(data=iris, x="species", y="sepal_width")
plt.title("Violin Plot of Sepal Width by Species")
plt.show()
```

8. Load dataset diamonds

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

9. Create a scatter plot with x-axis as 'carat' and y-axis as 'price' from the diamonds dataset.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=diamonds, x='carat', y='price', alpha=0.3)
plt.title("Carat vs Price")
plt.show()
```

10. Create a histogram of the 'price' column with 30 bins. Add a line that represents the mean price in red.

```python
plt.figure(figsize=(8,6))
sns.histplot(diamonds['price'], bins=30, kde=False)
plt.axvline(diamonds['price'].mean(), color='red', linestyle='--', linewidth=2, label='Mean Price')
plt.title("Histogram of Diamond Prices with Mean Line")
plt.legend()
plt.show()
```

11. Create a bar plot of the count of diamonds for each 'cut' category.

```python
plt.figure(figsize=(6,4))
sns.countplot(data=diamonds, x='cut')
plt.title("Count of Diamonds by Cut")
plt.show()
```

12. Create a box plot of the 'price' column for each 'cut' category.

```python
plt.figure(figsize=(8,6))
corr = diamonds.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

14. Create a line plot of the average 'price' per 'carat' for each 'cut' category.

```python
plt.figure(figsize=(8,6))
sns.lineplot(data=diamonds, x="carat", y="price", hue="cut", estimator="mean")
plt.title("Average Price per Carat by Cut")
plt.show()
```

15. Create a stacked bar chart of the count of diamonds by cut and color.

```python
import pandas as pd

# Create cross-tab
cut_color_counts = pd.crosstab(diamonds['cut'], diamonds['color'])

# Plot stacked bar chart
cut_color_counts.plot(kind='bar', stacked=True, figsize=(8,6))
plt.title("Stacked Bar Chart of Diamonds by Cut and Color")
plt.ylabel("Count")
plt.show()
```

16. Create a histogram of the carat weight distribution of diamonds with different color grades.

```python
plt.figure(figsize=(8,6))
sns.histplot(data=diamonds, x="carat", hue="color", bins=30, element="step", multiple="stack")
plt.title("Carat Weight Distribution by Color Grade")
plt.show()
```
